```python
import pandas as pd 
FULL=pd.read_excel('G:\AIMS_Lab\my_project\合金（下）\屈服强度/used_index.xlsx')
df=FULL[['Al fraction', 'Zn fraction', 'Mn fraction', 'Mg fraction', 'Li fraction', 'Sn fraction', 'Ce fraction', 'La fraction', 'Ca fraction', 'Si fraction', 'Cu fraction', 'Ni fraction',
  'Fe fraction', 'Bi fraction', 'Ti fraction', 'Sr fraction', 'Sm fraction', 'Sc fraction', 'Gd fraction', 'Y fraction', 'Zr fraction', 'Nd fraction', 'Ag fraction', 'Yb fraction','Grain Size','Precipitate Distribution','Habit Plane']]
non_zero_counts = {}
for col in df.columns:
    non_zero_counts[col] = df[col].astype(bool).sum()

# 输出每列非零元素的个数
for col, count in non_zero_counts.items():
    print(f"Column {col}: {count} non-zero elements")
    
```
