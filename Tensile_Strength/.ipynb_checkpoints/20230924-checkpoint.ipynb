```python
import re
convert_content_dict = {'MG': "Mg", 'mg': "Mg"}
drop_content = []
convert_text = {"0,3Mn": "0.3Mn", "纯Mg": "1Mg", "AZ31": "3Al 1Zn 0.3Mn", "AM30": "3Al 0.25Mn", "WE43": "92.2Mg-4Y-3.3Gd-0.5Zr", "WE54": "1Mg-5Y-1.5Nd-2Gd-0.4Zr"}
element_sets = set()
def extract_elements_and_numbers(text):
    try:
        if text.lower().startswith("mg"):
            text = "1" + text
            # print(text)
        for k, v in convert_text.items():
            if k in text:
                newText = text.replace(k, v)
                print(text, "->", newText)
                text = newText
        elements = re.findall(r'[A-Za-z]+', text)  # 匹配元素
    except:
        print("!ERROR", text)
        return {}
    for ind, content in list(enumerate(elements))[::-1]:
        if content.lower() == "wt":
            elements.pop(ind)
        else:
            if content in drop_content:
                print("!ERROR content", content, text)
                return {}
            newContent = convert_content_dict.get(content, content)
            elements[ind] = newContent
            element_sets.add(newContent)
    numbers = re.findall(r'\d*\.?\d+', text)  # 匹配数字
    for ind, i in enumerate(numbers):
        numbers[ind] = float(i)
    if len(elements) != len(numbers):
        print(ind, elements, numbers, text, text.startswith("Mg"))
    return dict(zip(elements, numbers))

d = c['合金成分(wt.%)'].apply(extract_elements_and_numbers)
print(element_sets)
elementsd = pd.DataFrame.from_records(d.values)
elementsd.index = c.index
display(elementsd)
```

```python
# 处理屈服强度
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_d = c["屈服强度"].apply(transform_float)
typed_d
```

```python
# 处理抗拉强度
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_e = c["抗拉强度 （UTS）"].apply(transform_float)
typed_e
```

```python
# 处理延伸率
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_f = c["延伸率"].apply(transform_float)
typed_f
```

```python
# 处理晶粒尺寸
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            if x == "（见文献图4-4）":
                return 0
            print(x)
typed_g = c["晶粒尺寸"].apply(transform_float)
typed_g
```

```python
# 处理N 惯习面（基面，柱面，基面+柱面，杂质，无）
# 处理N 惯习面（基面，柱面，基面+柱面，杂质，无）
def transform_panel(x):
    chemical, panel = x
    targetPorp = set()
    if pd.isna(panel):
        if pd.isna(chemical):
            return "无"
        else:
            print(chemical)
        if isinstance(chemical, int):
            print("数字？", chemical)
            return "无"
        if "LPSO" in chemical:
            targetPorp.add("基面")
        if "γ" in chemical:
            targetPorp.add("基面")
        if "β" in chemical:
            targetPorp.add("柱面")
        if "Mg" in chemical:
            targetPorp.add("杂质")
        if not len(targetPorp):
            print("！判别为杂质", chemical)
            return "杂质"
        else:
            if chemical.startswith("β(Mg"):
                print("**纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "-> 排除杂质")
                targetPorp.remove("杂质")
            target = "+".join(sorted(list(targetPorp)))
            if chemical in ["Gd,Mg", "Y,Mg", "Y,Gd,Mg"]:
                print("**纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "-> 无")
                return "无"
            print("纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "->", target)
            return target
    if panel.strip() in ["基面", "柱面", "基面+柱面"]:
        return panel.strip()
    else:
        if panel in ["球形颗粒"]:
            print("纯惯习面推断|", chemical, panel, "->", "杂质")
            return "杂质"
        if "LPSO" in chemical:
            targetPorp.add("基面")
        if "γ" in chemical:
            targetPorp.add("基面")
        if "β" in chemical:
            targetPorp.add("柱面")
        if "Mg" in chemical:
            targetPorp.add("杂质")
        if "LPSO" in panel:
            targetPorp.add("基面")
        if "γ" in panel:
            targetPorp.add("基面")
        if "β" in panel:
            targetPorp.add("柱面")
        if "Mg" in panel:
            targetPorp.add("杂质")
        if not len(targetPorp):
            print("!!!!未处理", chemical, panel)
        else:
            target = "+".join(sorted(list(targetPorp)))
            print("综合推断| 化学式：", chemical, " 惯习面标记：", panel, "->", target)
            return target
typed_N = c[['析出相-化学式', '析出相-形貌（球形，片层，杆状）']].apply(transform_panel, axis=1)
typed_N.name = "惯习面"
display(typed_N)


# def transform_panel(x):
#     if pd.isna(x):
#         return "无"
#     else:
#         print(x)
#         return "杂质"
# typed_N = c["析出相-化学式"].apply(transform_panel)
# typed_N.name = "惯习面"
```
