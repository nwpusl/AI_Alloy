```python
b = a[['Unnamed: 0', '流程', '参数', '屈服强度',
       '抗拉强度 （UTS）', '延伸率', '基体', 'Unnamed: 10', '析出相',
       'Unnamed: 12', 'Unnamed: 13', '析出相尺寸', 'Unnamed: 15', 'Unnamed: 16',
       '析出相体积分数', '析出相分布', 'Ref.（doi）']]
b.columns = ['合金成分 （wt.%）', '工艺-流程', '工艺-参数', '屈服强度',
       '抗拉强度 （UTS）', '延伸率', '晶粒尺寸', '显微组织特征-基体-固溶度', '显微组织特征-析出相-化学式/成分',
       '显微组织特征-析出相-形貌（球形，片层，杆状）', '显微组织特征-析出相-惯习面（基面，柱面）', '显微组织特征-析出相尺寸-直径', '显微组织特征-析出相尺寸-长度', '显微组织特征-析出相尺寸-宽度',
       '显微组织特征-析出相体积分数', '析出相分布', 'Ref.（doi）']
c = b.iloc[1:, :]
display(c)
```

```python
import re
convert_content_dict = {'MG': "Mg", 'mg': "Mg", "zn": "Zn", }
drop_content = []
convert_text = {"5.8Y-1.6Zn-0.3Z": "5.8Y-1.6Zn-0.3Zr", "(VWQK21)": "", "(VWQK12)": "", "Nd-Zr": "Nd-1Zr", "(VK)": "", "WK": "", "VQK": "", "WQK": "", "VWQK21": "", "(G15K)": "", "(GZ151)": "", "(GZ151K)": "", "(GWZ1311K)": "", "(GZ151M)": "", "GWZ1311M": "", "(at.%": "", "GZ ": ""}
element_sets = set()
def extract_elements_and_numbers(text):
    try:
        if text.lstrip().lower().startswith("mg"):
            text = "1" + text
            # print(text)
        for k, v in convert_text.items():
            if k in text:
                newText = text.replace(k, v)
                print(text, "->", newText)
                text = newText
        elements = re.findall(r'[A-Za-z]+', text)  # 匹配元素
    except:
        print("!ERROR", text)
        return {}
    for ind, content in list(enumerate(elements))[::-1]:
        if content.lower() == "wt":
            elements.pop(ind)
        else:
            if content in drop_content:
                print("!ERROR content", content, text)
                return {}
            newContent = convert_content_dict.get(content, content)
            elements[ind] = newContent
            element_sets.add(newContent)
    numbers = re.findall(r'\d*\.?\d+', text)  # 匹配数字
    for ind, i in enumerate(numbers):
        numbers[ind] = float(i)
    if len(elements) != len(numbers):
        print(ind, elements, numbers, text, text.startswith("Mg"))
    return dict(zip(elements, numbers))

d = c['合金成分 （wt.%）'].apply(extract_elements_and_numbers)
print(element_sets)
elementsd = pd.DataFrame.from_records(d.values)
elementsd.index = c.index
display(elementsd)
```

```python
# 处理屈服强度
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_d = c["屈服强度"].apply(transform_float)
typed_d
```

```python
# 处理抗拉强度
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_e = c["抗拉强度 （UTS）"].apply(transform_float)
typed_e
```

```python
# 处理延伸率
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            print(x)
typed_f = c["延伸率"].apply(transform_float)
typed_f
```

```python
# 处理晶粒尺寸
replace_target = {
    "3.3 um(17.6%)+50um(77.5%)": 41.3573081,  # (3.3 * 17.6 + 50 * 77.5) / (17.6 + 77.5)
    "3.3 um(14.1%)+50um(79.2%)": 42.94244373,
    "3.3 um(4.7%)+50um(87.9%)": 47.62969762,
    "3.3 um(66.6%)+50um(24.6%)": 15.89671053,
    "1.3un+60um": 30.65,
    "DRXed 1.7, 42.7%": 12.1859,
    "DRXed 1.6, 31%": 14.296,
    "3.7,  95.4%+57,4.6%": 6.1518,
    "4.0, 90.3% +110, 9.7%": 14.282,
    "4.2, 82.5%+ 178, 17.5%": 34.615,
    "3.00（细晶）": 11.5,
    "0.88（细晶）": 10.44,
}
def transform_float(x):
    try:
        return float(x)
    except Exception as ex:
        try:
            return float(x.split("±")[0])
        except Exception as ex:
            if x in replace_target:
                print(x, "->", replace_target[x])
                x = replace_target[x]
            else:
                print("!!!", x)
typed_g = c["晶粒尺寸"].apply(transform_float)
typed_g
```

```python
# 处理N 惯习面（基面，柱面，基面+柱面，杂质，无）
def transform_panel(x):
    chemical, panel = x
    targetPorp = set()
    if pd.isna(panel):
        if pd.isna(chemical):
            return "无"
        if "LPSO" in chemical:
            targetPorp.add("基面")
        if "γ" in chemical:
            targetPorp.add("基面")
        if "β" in chemical:
            targetPorp.add("柱面")
        if "Mg" in chemical:
            targetPorp.add("杂质")
        if not len(targetPorp):
            print("！判别为杂质", chemical)
            return "杂质"
        else:
            if chemical.startswith("β(Mg"):
                print("**纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "-> 排除杂质")
                targetPorp.remove("杂质")
            target = "+".join(sorted(list(targetPorp)))
            if chemical in ["Gd,Mg", "Y,Mg", "Y,Gd,Mg"]:
                print("**纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "-> 无")
                return "无"
            print("纯化学式推断| 化学式：", chemical, " 惯习面标记：", panel, "->", target)
            return target
    if panel.strip() in ["基面", "柱面", "基面+柱面"]:
        return panel.strip()
    else:
        if panel in ["球形颗粒"]:
            print("纯惯习面推断|", chemical, panel, "->", "杂质")
            return "杂质"
        if "LPSO" in chemical:
            targetPorp.add("基面")
        if "γ" in chemical:
            targetPorp.add("基面")
        if "β" in chemical:
            targetPorp.add("柱面")
        if "Mg" in chemical:
            targetPorp.add("杂质")
        if "LPSO" in panel:
            targetPorp.add("基面")
        if "γ" in panel:
            targetPorp.add("基面")
        if "β" in panel:
            targetPorp.add("柱面")
        if "Mg" in panel:
            targetPorp.add("杂质")
        if not len(targetPorp):
            print("!!!!未处理", chemical, panel)
        else:
            target = "+".join(sorted(list(targetPorp)))
            print("综合推断| 化学式：", chemical, " 惯习面标记：", panel, "->", target)
            return target
typed_N = c[["显微组织特征-析出相-化学式/成分", "显微组织特征-析出相-惯习面（基面，柱面）"]].apply(transform_panel, axis=1)
typed_N.name = "惯习面"
display(typed_N)
```

```python
# 处理S 析出相分布
# 请注意，这一步需要预先知道judge type！
def transform_distribution(x):
    judgeType = ["晶粒", "晶界", "晶内"]
    if pd.isna(x):
        return "无"
    else:
        j = []
        for ind, i in enumerate(judgeType):
            if i in x:
                j.append(i)
        # print("!!!!!", j)
        return "+".join(j)
typed_S = c["析出相分布"].apply(transform_distribution)
typed_S.name = "析出相分布"
display(typed_S)
```

```python
# 处理ZZ 将原表的信息进行登记
def transform_distribution(x):
    return "20231106-" + str(x["index"]) + "-" + str(x['合金成分 （wt.%）'])
typed_ZZ = c.reset_index()[["index", '合金成分 （wt.%）']].apply(transform_distribution, axis=1)
typed_ZZ.name = "追踪编号"
typed_ZZ.index = c.index
typed_ZZ
```

```python
elements_ress = pd.concat([elementsd, typed_g, typed_d, typed_e, typed_f, typed_N, typed_S, typed_ZZ], axis=1)[(~elementsd.isnull().all(axis=1)) & (~typed_d.isnull() | ~typed_e.isnull() | ~typed_f.isnull())]
elements_resss = elements_ress.fillna(0)
display(elements_resss)
```
