```python
import pandas as pd
from openpyxl import load_workbook
from strength_calculation import calculate_Strength
import os
select_top_feature_num = 10

def get_qf_data(writer):
    qf_data = pd.read_excel('20240607FULL.xlsx', index_col=0)
    qf_new = calculate_Strength(qf_data)
    top_fea_qf = pd.read_excel('qf_final_feature_shap_values.xlsx')['Feature Name']
    qf_fea_ls = top_fea_qf.to_list()
    qf_top_features_name = qf_fea_ls[:select_top_feature_num]
    important_features_name = ['Grain Size', 'Habit Plane', 'Diameter of basal phase', 'Length of Basal phase',
                               'Width of basal phase', 'Fraction of basal phase', 'Diameter of prismatic phase', 'Length of prismatic phase', 'Width of prismatic phase',
                               'Fraction of prismatic phase', 'Distribution of precipitation', '屈服强度', '抗拉强度 （UTS）']
    qf_features_name = list(set(qf_top_features_name + important_features_name))
    length = len(qf_features_name)
    qf_final = qf_new[qf_features_name].copy()
    qf_final = pd.concat([qf_final.drop(columns=['屈服强度', '抗拉强度 （UTS）']), qf_new[['屈服强度', '抗拉强度 （UTS）']]], axis=1)
    qf_final.to_excel(writer, sheet_name=f'qf_{length}', index=False)

def get_kl_data(writer):
    kl_data = pd.read_excel('20240607FULL.xlsx', index_col=0)
    kl_new = calculate_Strength(kl_data)
    top_fea_kl = pd.read_excel('kl_final_feature_shap_values.xlsx')['Feature Name']
    kl_fea_ls = top_fea_kl.to_list()
    kl_top_features_name = kl_fea_ls[:select_top_feature_num]
    important_features_name =['Grain Size', 'Habit Plane', 'Diameter of basal phase', 'Length of Basal phase',
                               'Width of basal phase', 'Fraction of basal phase', 'Diameter of prismatic phase', 'Length of prismatic phase', 'Width of prismatic phase',
                               'Fraction of prismatic phase', 'Distribution of precipitation', '屈服强度', '抗拉强度 （UTS）']
    kl_features_name = list(set(kl_top_features_name + important_features_name))
    length = len(kl_features_name)
    kl_final = kl_new[kl_features_name].copy()
    kl_final = pd.concat([kl_final.drop(columns=['屈服强度', '抗拉强度 （UTS）']), kl_new[['屈服强度', '抗拉强度 （UTS）']]], axis=1)
    kl_final.to_excel(writer, sheet_name=f'kl_{length}', index=False)
    print('save to',length,'xlsx')


output_file = '20240607FULL_final.xlsx'
if not os.path.exists(output_file):
    pd.DataFrame().to_excel(output_file)
with pd.ExcelWriter(output_file, engine='openpyxl', mode='a',if_sheet_exists = 'replace') as writer:
    get_qf_data(writer)
    get_kl_data(writer)
```
