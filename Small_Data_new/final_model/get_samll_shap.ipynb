```python
import pandas as pd

# 源文件
input_file = "G:\AIMS_Lab_final\my_project\AAA-MgAlloy\AI_Alloy\Small_Data_new/final_model\qf/rf_feature_shap_values.xlsx"
# 输出的汇总结果文件
output_file = "qf_shap_summary.xlsx"

# 读取 Excel 文件并获取所有 Sheet 的名称
xlsx = pd.ExcelFile(input_file)
sheet_names = xlsx.sheet_names

# 用于存放最终结果的空 DataFrame
summary_df = pd.DataFrame(columns=["SheetName", "MeanAbsSHAP"])

# 逐个 sheet 处理
for sheet in sheet_names:
    # 读取该 sheet 的数据
    df = pd.read_excel(input_file, sheet_name=sheet)
    
    # 假设表格里两列正好叫做 "Original Value" 和 "SHAP Value"
    # 对 SHAP Value 列取绝对值再平均
    mean_abs_shap = df["SHAP Value"].abs().mean()
    
    # 把结果存到我们的汇总表里
    summary_df.loc[len(summary_df)] = [sheet, mean_abs_shap]

# 将汇总结果输出到新的 Excel
summary_df.to_excel(output_file, index=False)
print("各 Sheet 对应的平均绝对 SHAP 值已输出到:", output_file)

# 如果想直接在屏幕上查看，也可以 print(summary_df)
```
