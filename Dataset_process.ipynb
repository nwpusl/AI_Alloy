```python
# 首先比较一下数据集并没有交集
import pandas as pd
data_origin = pd.read_excel('Original_Data/20240627Mg_alloy.xlsx',sheet_name = 'Sheet1',header = 0)
print(data_origin.columns)
display(data_origin)

# 选出模型所使用的基础特征作为原始数据
data_train = data_origin[['No.','合金成分（wt.%）',  'Grain Size','Precipitate Distribution', 'Habit Plane', '屈服强度','抗拉强度 (UTS)',]]
display(data_train)

# 需要判断这里的数据和验证集的数据有没有重复的
data_validation = pd.read_excel('Original_Data/20240127Validation.xlsx',header = 0)
ls = ['屈服强度','抗拉强度 (UTS)']
df1 = data_validation[ls]
df2 = data_train[ls]
common_rows = pd.merge(df1, df2, on=ls, how='inner')

# 检查是否有重复的行
if not common_rows.empty:
    print("存在重复的数据:")
    print(common_rows)
else:
    print("没有重复的数据。")
    print("经验证，并不为同一个材料")
```

```python
# 删除抗拉强度不正常的
ls = ['屈服强度','抗拉强度 (UTS)']

# 删除屈服强度或者抗拉强度为0的行或者列
# method 1 : 使用drop，直接修改原本的dataframe
zero_num=len(data_train[data_train['屈服强度']==0].index)
if zero_num == 0:
    print('没有为0的屈服强度')
else:
    print(f'有{zero_num}个为0的副本')
data_train.drop(data_train[data_train['屈服强度']==0].index,inplace = True)
display(data_train)

# 需要去掉晶粒尺寸400以上的数据
print(data_train['Grain Size'].nlargest(10))

drop_num= len(data_train[data_train['Grain Size']>=400])

data_train.drop(data_train[data_train['Grain Size']>=400].index, inplace = True)
print(f'晶粒尺寸提出{drop_num}个样本后的结果：')
print(data_train['Grain Size'].nlargest(10))
display(data_train)

# 需要去掉两个特别的晶粒值
data_train.drop(data_train[data_train['Precipitate Distribution']=='晶粒'].index,inplace = True)
# 验证数据集是否有重合
```

```python
# 开始改变合金成分
import re

df = data_train
# List of elements to include as columns
elements = ['Al', 'Zn', 'Mn', 'Mg', 'Li', 'Sn', 'Ce', 'La', 'Ca', 'Si', 'Cu', 'Ni',
            'Fe', 'Bi', 'Ti', 'Sr', 'Sm', 'Sc', 'Gd', 'Y', 'Zr', 'Nd', 'Ag', 'Yb']

for element in elements:
       df[element] = 0.0


def extract_elements(compound):
       # Extract element and percentage pairs using regular expression
       matches = re.findall(r'(\d+\.?\d*)\s*([A-Za-z]+)', compound)
       element_dict = {elem: float(pct) for pct,elem in matches}
       return element_dict


# Apply extraction to each row
for i, row in df.iterrows():
       compound = re.sub(r'\s*\([^)]*\)', '', row['合金成分（wt.%）'])
       element_dict = extract_elements(compound)

       # Update the DataFrame with the extracted values
       for element, percentage in element_dict.items():
              if element in elements:
                     df.at[i, element] = percentage

       # Calculate the Mg percentage
       total_other_elements = sum(element_dict.values())
       # 第i行的'Mg'元素含量计算
       df.at[i, 'Mg'] = 100.0 - total_other_elements


# 为每个材料设置基本的序号

df['追踪编号'] =  df['合金成分（wt.%）']

# 使用字典来跟踪每个成分名的出现次数
composition_count = {}
no_counter = 1  # 初始化No列的数字

for index, row in df.iterrows():
    compound = row['合金成分（wt.%）']
    # 如果内容在成分字典中，说明编号需要加1，同时赋值这个编号给材料
    if compound in composition_count:
        composition_count[compound] += 1
        df.at[index, '追踪编号'] += '-' + str(composition_count[compound])
    else:
        composition_count[compound] = 1
    # 在追踪编号开头添加No列的数字
    df.at[index, '追踪编号'] = 'No' + str(no_counter) + '-' + df.at[index, '追踪编号']
    no_counter += 1

display(df)
```

```python
import pandas as pd
# 对析出相和惯习面进行编码
df = pd.read_excel('train_data_20240607original.xlsx')
# df.to_excel('20240607.xlsx')

# 惯习面编码
# 定义映射字典
Habit_Panel = {
    '无': 0,
    '1': 1,
    '1+柱面': 2,
    "1+基面": 3,
    "基面": 4,
    "基面+柱面":5, # 球形颗粒也把它当做杂质
    "柱面": 6
}

Precipitation_phase_distribution = {
    '无': 0,
    '晶界': 1,
    '晶内': 2,
    '晶界+晶内': 3,
    '晶内+晶界': 3
}

# 使用replace方法对数据进行编码
df['Habit Plane'] = df['Habit Plane'].replace(Habit_Panel)
df['Precipitate Distribution'] = df['Precipitate Distribution'].replace(Precipitation_phase_distribution)

# 显示更新后的DataFrame以验证替换是否正确
print(df[['Habit Plane', 'Precipitate Distribution']])


# 计算pymatgen特征
from pymatgen.core import Composition
# print(df)
elements = ['Al', 'Zn', 'Mn', 'Mg', 'Li', 'Sn', 'Ce', 'La', 'Ca', 'Si', 'Cu', 'Ni',
            'Fe', 'Bi', 'Ti', 'Sr', 'Sm', 'Sc', 'Gd', 'Y', 'Zr', 'Nd', 'Ag', 'Yb']
weightPercentage = df[elements]
# print(weightPercentage)
# 元素
# 从权重字典中提取组分信息,转换为元素序列
def to_comp(x):
    try:
        return Composition.from_weight_dict(x.to_dict())
    except Exception as ex:
        print(x, ex)

compSeries = weightPercentage.apply(to_comp, axis=1)
print(compSeries)


from matminer.featurizers.composition.composite import ElementProperty
from matminer.featurizers.composition.alloy import WenAlloys
from matminer.featurizers.composition.element import ElementFraction
from matminer.featurizers.composition.composite import Meredig
import pandas as pd
# 创建 DataFrame
from strength_calculation import calculate_Strength
df_comp = pd.DataFrame({"composition": compSeries})
# print('hello')
# 初始化特征扩展器
featurizer = ElementProperty.from_preset("magpie")
# 应用特征扩展器
df_comp = featurizer.featurize_dataframe(df_comp, "composition")
df_comp = WenAlloys().featurize_dataframe(df_comp, "composition")
df_comp = Meredig().featurize_dataframe(df_comp, "composition")
for columnName, obj in df_comp.dtypes.items():
    if obj == "object":
        print(columnName)
        del df_comp[columnName]
        
all = pd.concat([df_comp,df[['Grain Size', 'Precipitate Distribution',
       'Habit Plane', '屈服强度', '抗拉强度 (UTS)', '追踪编号']]],axis = 1)
all = calculate_Strength(all)
```
