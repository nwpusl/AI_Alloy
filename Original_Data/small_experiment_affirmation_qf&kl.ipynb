```python
# 开始训练出新的qf模型  rf
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import r2_score, mean_absolute_percentage_error
from sklearn.ensemble import RandomForestRegressor
from openpyxl import load_workbook
import joblib
# 设置随机种子以确保结果的可重复性
Regressor=RandomForestRegressor
model_name='rf'
problem_name='qf'
main_path=f'../origin_model_full/{problem_name}/{model_name}'
excel=f'../origin_model_full/{problem_name}/{model_name}/{model_name}_index.xlsx'

if not os.path.exists(main_path):
    os.makedirs(main_path)
if not os.path.exists(excel):
    pd.DataFrame().to_excel(excel)
    
n_splits=5
ls1=['fold'+str(i) for i in range(1,n_splits+1)]

fold_dict=dict(zip(ls1,[0 for i in range(0,len(ls1))]))

np.random.seed(200)
# 读取 Excel 文件
df = pd.read_excel('../FULL_784.xlsx', index_col=0)
# 删除包含空值的行
df.dropna(inplace=True)
df=df[df['屈服强度']!=0].reset_index(drop=True)

feature_names=df.columns[:-4]
# 分离特征和目标变量
X = df.iloc[:, :-4]  # 特征: 最后四列之前的所有列
# print('X:',X)

print('特征的形状:\n',len(feature_names))
print('输入数据的形状：\n',X.shape)
y = df['屈服强度']  # 目标: 倒数第四列

# print('y:',y)
# 将目标变量分箱为10个类别
bins = np.linspace(y.min(), y.max(), 11)
y_binned = np.digitize(y, bins)

# 初始化StratifiedKFold
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=200)

params = {
    'bootstrap': True,
    'ccp_alpha': 0.0,
    'criterion': 'squared_error',
    'max_depth': None,
    'max_features': 1.0,
    'max_leaf_nodes': None,
    'max_samples': None,
    'min_impurity_decrease': 0.0,
    'min_samples_leaf': 1,
    'min_samples_split': 2,
    'min_weight_fraction_leaf': 0.0,
    'n_estimators': 100,
    'n_jobs': None,
    'oob_score': False,
    'random_state': 200,
    'verbose': 0,
    'warm_start': False
}

# 初始化随机森林回归器，最大深度为40，预测器（树）数量为100
model= Regressor(**params)  # 如果这里固定初始化模型删除掉的话可以在下面循环呢内添加随机初始化或者指定个别参数

# 初始化列表以存储每个折叠的分数
r2_scores = []
mape_scores = []

# 进行5折交叉验证
index=0
num_each_fold_train=np.floor((np.int32(X.shape[0])-1)*(n_splits-1)/n_splits)
num_each_fold_test=(np.int32(X.shape[0])-1)//n_splits
for train_index, test_index in skf.split(X, y_binned):
    X_train, X_test = X.iloc[train_index[:int(num_each_fold_train)]], X.iloc[test_index[:int(num_each_fold_test)]]  #强制转换为整形
    y_train, y_test = y.iloc[train_index[:int(num_each_fold_train)]], y.iloc[test_index[:int(num_each_fold_test)]]
    index=index+1
    fold_dict['fold' + str(index)] = train_index[:int(num_each_fold_train)].astype(int)

    # 训练模型
    model.fit(X_train, y_train)

    # 在测试集上进行预测
    y_pred = model.predict(X_test)
    # 计算并存储分数
    r2_scores.append(r2_score(y_test, y_pred))  # 计算拟合优度
    print(r2_scores[-1])
    mape_scores.append(mean_absolute_percentage_error(y_test, y_pred))  # 获得平均绝对百分比误差

    # 保存模型文件
    joblib.dump(model,   f'{main_path}/{model_name}{index}.pkl')
    print('保存模型文件')

    feature_importance_final = model.feature_importances_
#     print("importance_shpe:",(feature_importance_final).size)
    importance_dict = dict(zip(feature_importance_final, feature_names))
    importance_tuplelist = [(abs(value), feature) for (value, feature) in importance_dict.items()]
    importance_sorted = sorted(importance_tuplelist, reverse=True)

    # 输出排名靠前列表的序号
    x_feature = [j for (i, j) in importance_sorted]
    y_importance_value = [i for (i, j) in importance_sorted]
#     print('rf_tree_feature_names(top15):')
#     for i in x_feature[:15]:
#         print(i)
#     print(f'average_xgboost_feature_values(top15):')
#     for i in y_importance_value[:15]:
#         print(i)

ls = list(fold_dict.values())
# print(ls)
df = pd.DataFrame(ls)
book = load_workbook(excel)
with pd.ExcelWriter(excel) as writer:
    writer.book = book
    df.to_excel(writer, sheet_name='k-fold')

# df = pd.read_excel(excel, sheet_name='k-fold')
# print('训练集序号：', list(df))

# 打印平均分数
print('平均 R2 分数:', np.mean(r2_scores))
print('平均 MAPE 分数:', np.mean(mape_scores) * 100)  # 转换为百分比
```
