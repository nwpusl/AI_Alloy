```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
def scores(pred, target):
    pred = np.array(pred)
    target = np.array(target)
    mae = mean_absolute_error(pred, target)
    mse = mean_squared_error(pred,target)
    mape = mean_absolute_percentage_error(pred,target)
    r2 = r2_score(pred,target)
    return {'mae':mae,"mse":mse,"mape":mape,"r2":r2}

# 示例数据
data = pd.read_excel("../../Yield_Strength/test_set_new.xlsx",header=0)
df = pd.DataFrame(data)
df.columns = df.columns.str.replace(' ', '_')

# 定义表达式
# expression = "((((Calculated_Yield_Strength * Grain_Size) * 0.023342336) + Calculated_Yield_Strength) * (Calculated_Yield_Strength * -0.00092431164)) + (Calculated_Yield_Strength + 76.30307)"
expression = "(frac_f_valence_electrons + 0.122380376) * (786.8776 - (-18033.59 / (Grain_Size + 13.1522255)))"
# replace_strength = expression.replace('_',' ')

# 将表达式中的变量替换为 DataFrame 列的引用
for col in df.columns:
    expression = expression.replace(col, f"df['{col}']")

# print(df.columns)

# 计算结果
df['Result'] = eval(expression)
res = scores(df['Result'],df['屈服强度'])
# print(df['Result'])
print('pred:',res)
calculation_imprical = scores(df['Calculated_Yield_Strength'],df['屈服强度'])
print('impirical:',calculation_imprical)
```
